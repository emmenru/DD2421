# LABELING STUFF

```python
# Load existing progress if it exists, otherwise start fresh
progress_file = Path('/content/drive/MyDrive/G3S-Univ-Paris-8/Colabs/screening_progress_combined.csv')

if progress_file.exists():
    df_clean = pd.read_csv(progress_file)
    print("Loaded existing progress")
else:
    # Only initialize columns if starting fresh
    df_clean[['obj1', 'obj2', 'obj3', 'exclude']] = None
    print("Starting fresh")

# Need to set starting point so that you can restart where you stopped
current_index = [1900]

output = Output()

def show_paper():
    with output:
        clear_output(wait=True)
        i = current_index[0]
        if i >= len(df_clean):
            print("All papers reviewed!")
            print(f"Included Objective 1: {(df_clean['obj1'].notna()).sum()}")
            print(f"Included Objective 2: {(df_clean['obj2'].notna()).sum()}")
            print(f"Included Objective 3: {(df_clean['obj3'].notna()).sum()}")
            print(f"Excluded: {(df_clean['exclude'].notna()).sum()}")
            return

        print(f"Paper {i+1}/{len(df_clean)}")
        display(df_clean.loc[[i],['ID', 'title', 'keywords', 'abstract']])

def on_decision(decision):
    index = current_index[0]

    if decision == 'exclude':
        # Clear all objectives and set exclude
        df_clean.loc[index, ['obj1', 'obj2', 'obj3']] = None
        df_clean.loc[index, 'exclude'] = 'exclude'
    elif decision == 'increment':
        # Check if current paper has been annotated
        has_annotation = df_clean.loc[index, ['obj1', 'obj2', 'obj3', 'exclude']].notna().any()

        if not has_annotation:
            with output:
                print("Please annotate this paper before moving to the next one!")
            return

        # Save before incrementing
        df_clean.to_csv('/content/drive/MyDrive/G3S-Univ-Paris-8/Colabs/screening_progress_combined.csv', index=False)
        # Then augment counter
        current_index[0] += 1
        show_paper()
        return
    else:
        # Clear exclude and set the selected objective
        df_clean.loc[index, 'exclude'] = None
        df_clean.loc[index, decision] = decision

    # Save after each decision
    df_clean.to_csv('/content/drive/MyDrive/G3S-Univ-Paris-8/Colabs/screening_progress_combined.csv', index=False)

# Create buttons with configuration
button_configs = {
    'obj1': {'color': 'lightgreen'},
    'obj2': {'color': 'lightblue'},
    'obj3': {'color': 'lightpink'},
    'done': {'color': 'yellow', 'action': 'increment'},
    'exclude': {'action': 'exclude', 'color': 'red'}
}

buttons = {}
for name, config in button_configs.items():
    btn = Button(description=name)

    if 'color' in config:
        btn.style.button_color = config['color']

    action = config.get('action', name)
    btn.on_click(lambda b, action=action: on_decision(action))

    buttons[name] = btn

display(HBox([buttons['obj1'], buttons['obj2'], buttons['obj3'], buttons['exclude'], buttons['done']]))
display(output)
show_paper()
```

```python
# Make sure exclude column is int and the others are categories

df_clean['obj1'] = df_clean['obj1'].astype("category")
df_clean['obj2'] = df_clean['obj2'].astype("category")
df_clean['obj3'] = df_clean['obj3'].astype("category")
df_clean['exclude'] = df_clean['exclude'].astype("category")
```
