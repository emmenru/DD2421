````python
# Make sure we have the normalized column
df['Author Keywords Normalized'] = df['Author Keywords'].str.lower()

# Get all unique keywords sorted by frequency
all_keywords_sorted = keywords.value_counts().index.tolist()

@interact(
    keyword=Dropdown(options=['-- All Papers --'] + all_keywords_sorted,
                     value='-- All Papers --',
                     description='Keyword:'),
    max_rows=IntSlider(min=10, max=100, step=10, value=50, description='Max rows:')
)
def filter_papers(keyword='-- All Papers --', max_rows=50):
    if keyword == '-- All Papers --':
        filtered_df = df
        print(f"Showing all {len(filtered_df)} papers (displaying first {max_rows})\n")
    else:
        # Split keywords, strip whitespace, and check for exact match
        def has_keyword(x):
            if pd.notna(x):
                kw_list = [kw.strip() for kw in str(x).split(';')]
                return keyword in kw_list
            return False

        mask = df['Author Keywords Normalized'].apply(has_keyword)
        filtered_df = df[mask]
        print(f"Papers containing '{keyword}': {len(filtered_df)} found\n")

    display_cols = ['Title', 'Year', 'Author full names', 'Author Keywords']
    display_df = filtered_df[display_cols].head(max_rows)

    # Style the dataframe for better readability
    styled_df = display_df.style.set_properties(**{
        'text-align': 'left',
        'white-space': 'pre-wrap',
        'word-wrap': 'break-word',
        'max-width': '400px'
    })

    return styled_df
````
